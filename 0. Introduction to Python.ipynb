## Loops

```python
for element in different_element_list:
  print(element)
```

```python
for i in range(len(different_element_list)):
  print(i, "-->", different_element_list[i])
```

Loop with step

```python
for i in range(0, len(different_element_list), 2):
  print(i, "-->", different_element_list[i])
```

```python
i = 0
while i < len(different_element_list):
  print(different_element_list[i])
  i += 1
```

# 5. Numpy and Pandas

NumPy is a Python library used for working with arrays.

It also has functions for working in domain of linear algebra, fourier transform, and matrices.

NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.

This behavior is called locality of reference in computer science.

This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

```python
import numpy as np
import pandas as pd
```
